# Data Preprocessing

```python
Claimants_Data["CLMSEX"]=Claimants_Data["CLMSEX"].fillna(1)
```

```python
Claimants_Data.isnull().sum()
```

```python
Claimants_Data["CLMINSUR"]=Claimants_Data["CLMINSUR"].fillna(1)
```

```python
Claimants_Data.isnull().sum()
```

```python
Claimants_Data["SEATBELT"]=Claimants_Data["SEATBELT"].fillna(1)
```

```python
Claimants_Data.isnull().sum()
```

```python
Claimants_Data["CLMAGE"].mean() 
```

```python
Claimants_Data["CLMAGE"]=Claimants_Data["CLMAGE"].fillna(Claimants_Data["CLMAGE"].mean())# if the no of data increases, the mean will be caluculated automatically and filled
```

```python
Claimants_Data.isnull().sum()
```

# Data preparation

## Separate x and y

```python
x=Claimants_Data.drop(labels="ATTORNEY",axis=1)
x
```

```python
y=Claimants_Data[["ATTORNEY"]]
y
```

* Again spliting the x as a part to train, as a part to test

* spliting y as a part to train and as a part to test


## Model validation technique

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y,  train_size=0.80,random_state=45,
    shuffle=True)
```

```python
X_train #80% data is used for training
```

```python
X_test #balance 20% is used for testing data is shuffled
```

```python
y_train
```

```python
y_test
```

# Model Building

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

# Model Prediction

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score
```

```python
y_train_pred=model.predict(X_train)
```

```python
accuracy_score(y_train, y_train_pred)
```

```python
y_test_pred=model.predict(X_test)
```

```python
accuracy_score(y_test, y_test_pred)
```

```python
confusion_matrix(y_train, y_train_pred)
```

```python
confusion_matrix(y_test, y_test_pred)
```

```python
precision_score(y_test, y_test_pred)
```

```python
recall_score(y_test, y_test_pred)
```

```python
recall_score(y_train, y_train_pred)
```
