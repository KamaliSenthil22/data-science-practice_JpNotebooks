#### 2.Which age group has the highest count of people who sleeps less than 5 hours

#### 3.Who has the highest anxiety level

```python
Phone_addiction.sort_values("Anxiety_Level",ascending = False).head()
```

#### 4.Maximum number of apps used daily

```python
Max_no_apps_used =Phone_addiction["Apps_Used_Daily"].sort_values(ascending=False).head(1)
```

```python
pd.Series(Max_no_apps_used)
```

#### 5.which age group has less self esteem

```python
Avg_self_esteem=Phone_addiction.groupby("Age")["Self_Esteem"].mean()
lowest_group = Avg_self_esteem.idxmin()
lowest_value = Avg_self_esteem.min()
print("Age group with lowest self-esteem:", lowest_group)
print("Average self-esteem:", lowest_value)
```

```python
Num_cols=[ 'Age','Daily_Usage_Hours', 'Sleep_Hours', 'Academic_Performance',
       'Social_Interactions', 'Exercise_Hours', 'Anxiety_Level',
       'Depression_Level', 'Self_Esteem', 'Parental_Control',
       'Screen_Time_Before_Bed', 'Phone_Checks_Per_Day', 'Apps_Used_Daily',
       'Time_on_Social_Media', 'Time_on_Gaming', 'Time_on_Education', 'Family_Communication', 'Weekend_Usage_Hours',
       'Addiction_Level']
for col in Num_cols:
    if col in Phone_addiction.columns:
        plt.figure(figsize=(8,4))
        sns.histplot(Phone_addiction[col].dropna(),kde= True, bins=30)
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.tight_layout() # Neat spacing
        plt.show()
```

```python
# Countplot for categorical variable: Gender
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', data=Phone_addiction)
plt.title('Count of Gender')
plt.tight_layout()
plt.show()
```

```python
# Box Plot to see relationship between Daily Usage Hours and Addiction Level
plt.figure(figsize=(8,4))
sns.boxplot(x='Addiction_Level', y='Daily_Usage_Hours', data=Phone_addiction)
plt.title('Daily Usage Hours vs Addiction Level')
plt.tight_layout()
plt.show()
```
