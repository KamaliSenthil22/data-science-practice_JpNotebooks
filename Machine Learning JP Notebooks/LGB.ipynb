```python
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,AdaBoostClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
x=data.drop(labels="DEFAULT",axis=1)
```

```python
x
```

```python
y=data["DEFAULT"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.80)
```

```python
RF_model=RandomForestClassifier()
AG_model=AdaBoostClassifier()
G_mdoel=GradientBoostingClassifier()
XGB_model=XGBClassifier()
LGB_model=LGBMClassifier()
```

```python
%%time 
RF_model.fit(x_train,y_train)
```

```python
y_pred=RF_model.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
%%time 
AG_model.fit(x_train,y_train)
```

```python
y_pred=AG_model.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
%%time 
G_mdoel.fit(x_train,y_train)
```

```python
y_pred=G_mdoel.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
%%time 
XGB_model.fit(x_train,y_train)
```

```python
%%time 
LGB_model.fit(x_train,y_train)
```

```python
RF_Feature=RF_model.feature_importances_

AG_Feature=AG_model.feature_importances_

G_Feature=G_mdoel.feature_importances_

XGB_Feature=XGB_model.feature_importances_

LGB_Feature=LGB_model.feature_importances_
```

```python
LGB_Feature
```

```python
feature_importance=pd.DataFrame({"Feature":x_train.columns,"RF_Feature":RF_Feature,"AG_Feature":AG_Feature,"G_Feature":G_Feature,"XGB_Feature":XGB_Feature,"LGB_Feature":LGB_Feature})
```

```python
feature_importance
```

```python
Lgb_imp=feature_importance.sort_values(by="LGB_Feature",ascending=False)
```
