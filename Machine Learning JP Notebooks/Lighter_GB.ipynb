```python
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier,RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
x=credit_card_df.drop(labels="DEFAULT",axis=1)
x
```

```python
y=credit_card_df[["DEFAULT"]]
y
```

```python
 X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.80, )
```

```python
RF_model=RandomForestClassifier()
ADGB_model=AdaBoostClassifier()
GB_model=GradientBoostingClassifier()
XGB_model=XGBClassifier()
LGB_model=LGBMClassifier()
```

```python
%%time
RF_model.fit(X_train,y_train)
```

```python
y_predict=RF_model.predict(X_test)
```

```python
accuracy_score(y_test,y_predict)
```

```python
%%time
ADGB_model.fit(X_train,y_train)
```

```python
y_predict=ADGB_model.predict(X_test)
accuracy_score(y_test,y_predict)
```

```python
%%time
GB_model.fit(X_train,y_train)
```

```python
y_predict=GB_model.predict(X_test)
accuracy_score(y_test,y_predict)
```

```python
%%time
XGB_model.fit(X_train,y_train)
```

```python
y_predict=XGB_model.predict(X_test)
accuracy_score(y_test,y_predict)
```

```python
%%time
LGB_model.fit(X_train,y_train)
```

```python
y_predict=LGB_model.predict(X_test)
accuracy_score(y_test,y_predict)
```

## Feature importance variable

```python
RF_Feature=RF_model.feature_importances_

ADGB_Feature=ADGB_model.feature_importances_

GB_Feature=GB_model.feature_importances_

XGB_Feature=XGB_model.feature_importances_

LGB_Feature=LGB_model.feature_importances_
```

```python
LGB_Feature
```

```python
X_train.columns
```

```python
feature_importance=pd.DataFrame({"Feature":X_train.columns,"RF_Feature":RF_Feature,"ADGBG_Feature":ADGB_Feature,"GB_Feature":GB_Feature,"XGB_Feature":XGB_Feature,"LGB_Feature":LGB_Feature})
```

```python
feature_importance
```

```python
Lgb_imp=feature_importance.sort_values(by="LGB_Feature",ascending=False)
Lgb_imp
```

```python
import seaborn as sns 
sns.barplot(x="LGB_Feature",y="Feature",data=Lgb_imp)
```

```python
RF_Feature_imp=feature_importance.sort_values(by="RF_Feature",ascending=False)
sns.barplot(x="RF_Feature",y="Feature",data=RF_Feature_imp)
```
