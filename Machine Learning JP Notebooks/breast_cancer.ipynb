```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
from matplotlib import pyplot as plt
```

```python
cancer_data=load_breast_cancer()
```

```python
cancer_data.keys()
#"data" - x
#"traget" - y
#"feature_names" - columns name
# using pandas library the data is converted in dataframe
```

```python
x=pd.DataFrame(data=cancer_data["data"],
    columns=cancer_data['feature_names'])
x
```

```python
y=pd.DataFrame(data=cancer_data['target'])
y
```

### Now x and y is ready

# Model Validation Technique

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.20, random_state=4)
```

```python
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.20, random_state=4,stratify=y)
```

```python
plt.pie(y_train.value_counts(),autopct="%0.1f%%")
```

```python
plt.pie(y_test.value_counts(),autopct="%0.1f%%")
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score
```

```python
model=AdaBoostClassifier(n_estimators=100)
#estimator=None by defalut it will create desicion tree model if estimator is not mentioned
#n_estimator=100, we use 100 desicion tree in a sequential model
```

```python
model.fit(x_train,y_train)
```

```python
y_train_pred=model.predict(x_train)
```

```python
y_test_pred=model.predict(x_test)
```

```python
accuracy_score(y_train,y_train_pred)
```

```python
accuracy_score(y_test,y_test_pred)
```
