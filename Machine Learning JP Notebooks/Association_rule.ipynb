```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
OnlineRetail_df=pd.read_excel("Online Retail.xlsx")
```

# EDA

```python
OnlineRetail_df.shape
```

```python
France_df=OnlineRetail_df[OnlineRetail_df["Country"] == "France"]
```

```python
France_df
```

```python
France_df.isnull().sum()
```

```python
df_france=pd.pivot_table(data=France_df,values="Quantity",index="InvoiceNo",columns="Description").fillna(0)
df_france
```

```python
encoded_france=df_france.applymap(lambda x : 1 if x>0 else 0)
encoded_france
```

```python
pip install mlxtend
```

```python
frequent_item=apriori(encoded_france,
    min_support=0.05,
    use_colnames=True)
frequent_item
```

```python
Best_association=association_rules(frequent_item)
Best_association
```
