# Feature Engineering
* Feature Elimination - Dropping the unwanted feature
* Feature addition - adding some new features
* Feature selction- Identity the important feature and try to select those features

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
```

```python
Cancer_data=load_breast_cancer()
```

```python
Cancer_data
```

```python
x=pd.DataFrame(data=Cancer_data.data,columns=Cancer_data.feature_names)

x
```

```python
y=Cancer_data.target
```

# Two types of Feature Engineering
* Select from Model
* Recursive feature elimination

# Model validation technique

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.80, random_state=42)
```

```python
from sklearn.feature_selection import SelectFromModel
```

```python
select_from_model=SelectFromModel(estimator=RandomForestClassifier(n_estimators=50,
    criterion='gini',
    max_depth=10))
```

```python
select_from_model.fit(x_train,y_train)
```

```python
select_from_model.get_support()
```

```python
x.columns[select_from_model.get_support()]
```

```python
x_train_transformed=select_from_model.transform(x_train)
x_test_transformed=select_from_model.transform(x_test)
```

```python
x_test_transformed.shape
```

```python
x_train_transformed.shape
```

```python
def run_randomforest(x_train, x_test, y_train, y_test ):
    rf_model=RandomForestClassifier(max_depth=7)
    rf_model.fit(x_train,y_train)
    y_pred=rf_model.predict(x_test)
    print("accuracy_score:", accuracy_score(y_test,y_pred))
```

```python
run_randomforest(x_train, x_test, y_train, y_test)
```

```python
run_randomforest(x_train_transformed,x_test_transformed, y_train, y_test)
```

### In the select from model number of features selection is not in our control

### But in Recursive feature elimination, we can control the feature elimination

```python
from sklearn.feature_selection import RFE
```

```python
RFE_model=RFE(estimator=RandomForestClassifier(n_estimators=50,
    criterion='gini',
    max_depth=10))
```

```python
RFE_model.fit(x_train,y_train)
```

```python
RFE_model.get_support()
```

```python
x.columns[RFE_model.get_support()]
```

```python
x.columns[RFE_model.get_support()].shape
```

```python
x_train_transformed_rfe=RFE_model.transform(x_train)
x_test_transformed_rfe=RFE_model.transform(x_test)
```

```python
run_randomforest(x_train_transformed_rfe,x_test_transformed_rfe, y_train, y_test)
```

```python
for i in range(1,len(x.columns)):
    RFE_model=RFE(estimator=RandomForestClassifier(n_estimators=50,criterion='gini',max_depth=10),n_features_to_select=i)
    RFE_model.fit(x_train,y_train)
    
    x_train_transformed_rfe=RFE_model.transform(x_train)
    x_test_transformed_rfe=RFE_model.transform(x_test)

    print("Number of Features:",i)
    run_randomforest(x_train_transformed_rfe,x_test_transformed_rfe, y_train, y_test)
    print("#####################################################")
```
