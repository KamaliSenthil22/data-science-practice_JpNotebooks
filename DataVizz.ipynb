# Data visualization

## 1. Plot a line graph showing sales over 6 months.
* Months: ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
* Sales: [250, 270, 300, 280, 320, 340]


```python
import pandas as pd
from matplotlib import pyplot as plt
Months= ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
Sales= [250, 270, 300, 280, 320, 340]
```

```python
plt.plot(Months,Sales,marker='o', color='blue', linestyle='-')
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly Sales")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 2. Create a bar plot showing average marks of students in 3 subjects using Seaborn.
* Subjects: ['Math', 'Science', 'English']
* Marks: [80, 70, 90]

```python
import seaborn as sns
Subjects= ['Math', 'Science', 'English']
Marks= [80, 70, 90]
df = pd.DataFrame({
    'Subjects': Subjects,
    'Marks': Marks
})
df
```

```python
sns.barplot(x='Subjects', y='Marks', data=df, hue='Subjects', palette='pastel', legend=False)

plt.xlabel("Subjects")
plt.ylabel("Average_Marks")
plt.title("Average marks in Subjects")
plt.ylim(0,100)
plt.tight_layout()
plt.show()
```

## 3. Create a heatmap showing correlation between columns in the following



```python
Data=pd.DataFrame({
    "Age":[25,30,35,40],
    "Salary":[40000,50000,60000,80000],
    "Experience":[2,4,6,9]})
Data
```

```python
corr_report=Data.corr()
corr_report
```

```python
sns.heatmap(data=corr_report,annot=True)
plt.title("Correlation Heatmap of Age, Salary, and Experience")
plt.show()
```
