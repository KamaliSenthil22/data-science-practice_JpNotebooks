## 2. Plot a histogram of 5000 samples drawn from an exponential distribution

```python
from matplotlib import pyplot as plt #
import numpy as np

Samples_1=np.random.exponential(scale=2.0, size=5000) #Time between random events, scale : Average time between events (mean)
plt.hist(Samples_1,bins=50,color="orange",edgecolor='black')
plt.title("Histogram of Exponential Distribution (5000 Samples)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

## 3. Write a function to compute the expected value of a discrete random variable given values and probabilities.

```python

def expected_value(values,probabilities):
    if len(values)!=len(probabilities):
        return("Error: Lengths don't match")
    return sum(val*prob for val,prob in zip(values,probabilities))
```

```python
values = [1, 2, 3, 4]
probabilities = [0.1, 0.2, 0.3, 0.4]
expected_value(values,probabilities)
```

## 4. Plot and compare histograms of samples from normal and uniform distributions with same size.sssss

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 samples
normal_samples = np.random.normal(loc=0, scale=1, size=1000)  # mean = 0, std = 1
uniform_samples = np.random.uniform(low=-3, high=3, size=1000)  # equal spread from -3 to 3

# Plot histograms
plt.hist(normal_samples, bins=30, alpha=0.5, label='Normal', color='blue')#bins:It splits your data range into 30 equal-width intervals.
plt.hist(uniform_samples, bins=30, alpha=0.5, label='Uniform', color='orange')#alpha=Transparency of the bars.
plt.legend()#Displays a legend using the label= names provided earlier ('Normal', 'Uniform').
plt.title('Normal vs Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

Normal distribution will look like a bell curve

Uniform distribution will look flat or rectangular

You can visually compare how values are spread in both
